```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dirty_iris.csv")
df
```

```python
num = df._get_numeric_data()          #shallow copy
num[num < 0] = np.nan
num[num  == np.inf] = np.nan
print(df)
```

```python
f = df.dropna()
number_complete = len(f)
print("Number of complete rows: ",number_complete)
print("Percentage of complete rows: ",number_complete / len(df) * 100)
```

```python
def check_species(df):
    species = set(df.Species)
    func = lambda x : x in species
    plt_df =  df["Species"].apply(func)
    return plt_df
check_species(df)
```

```python
def all_positive(df):
    func = lambda x : x > 0
    func1= lambda x: x[0] and x[1] and x[2] and x[3]
    plt_df =  df.iloc[:,:-1].apply(func).apply(func1,axis=1)
    pl = pd.DataFrame(plt_df).apply(pd.value_counts)
    pl.plot.bar()
    return plt_df
print(all_positive(df))
```

```python
def length(df):
    func = lambda x : x[0] >= 2 * x[1]
    plt_df = df[["Petal.Length","Petal.Width"]].apply(func,axis = 1)
    pl = pd.DataFrame(plt_df).apply(pd.value_counts)
    pl.plot.bar()
    return plt_df
length(df)
```

```python
def sepal_length_lessthan30(df):
    func = lambda x : x < 30
    plt_df = df["Sepal.Length"].apply(func)
    pl = pd.DataFrame(plt_df).apply(pd.value_counts)
    pl.plot.bar()
    return plt_df
sepal_length_lessthan30(df)
```

```python
def sepal_longer_than_petal(df):
    func = lambda x : x[0] > x[1]
    plt_df = df[["Sepal.Length","Petal.Length"]].apply(func,axis = 1)
    pl = pd.DataFrame(plt_df).apply(pd.value_counts)
    pl.plot.bar()
    return plt_df
sepal_longer_than_petal(df)
```
